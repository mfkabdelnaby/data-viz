# Bokeh
Bokeh is an interactive visualization library that targets modern web browsers for presentation.

### Glyph

The basic visual building blocks of Bokeh plots, e.g. lines, circles, rectangles, squares, wedges, patches, etc. The `bokeh.plotting` interface provides a convenient way to create plots centered around glyphs.

Types of glyphs available in Bokeh:

`asterisk()`
`circle()`
`circle_cross()`
`circle_x()`
`cross()`
`diamond()`
`diamond_cross()`
`inverted_triangle()`
`square()`
`square_cross()`
`square_x()`
`triangle()`
`x()`

and others you can find in the documentation

````python
# imports

from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook

from bokeh.plotting import ColumnDataSource
from bokeh.layouts import row, column
````

````python
import numpy as np
import pandas as pd
````

````python
# data to plot

fertility = [1.769,2.682,2.077,2.140, 1.827]
female_literacy = [90.5,50.8,99.0,88.8,90.2]
````

````python
# Create the figure with its properties: p
p = figure(plot_width=400, plot_height=400,
           x_axis_label='fertility (children per woman)',
           y_axis_label='female_literacy (% population)')

# Add a circle glyph to the figure p
p.circle(fertility, female_literacy)

# Call the output_notebook() function to diplay it in notebook
output_notebook()

# Display the plot
show(p)
````

### Customizing your scatter plots

* **`color`**

Bokeh accepts [colors as hexadecimal strings](https://en.wikipedia.org/wiki/Web_colors), tuples of RGB values between 0 and 255, and any of the 147 [CSS color names](http://www.colors.commutercreative.com/grid/)

* **`size`**

Size values are supplied in screen space units with 100 meaning the size of the entire figure.

* **`alpha`**

Controls transparency. It takes in floating point numbers between 0.0, meaning completely transparent, and 1.0, meaning completely opaque.

````python
fertility2 = [2.26, 2.156, 2.404,2.223,2.53]
female_literacy2 = [94.9, 91.5, 93.4, 97.7, 84.6]
````

````python
# you can add any property of a figure
p.x(fertility2, female_literacy2, color='red', size=10,
   alpha=0.6)

output_notebook()
show(p)
````

### Patches

In Bokeh, extended geometrical shapes can be plotted by using the `patches()` glyph function. The patches glyph takes as input a list-of-lists collection of numeric values specifying the vertices in x and y directions of each distinct patch to plot.

Patches are useful in many applications. For instance, they can be used to shade geometric regions in a map.

````python
coordinates = pd.read_csv('coordinates.csv')
coordinates.head()

x = []
for colx in coordinates.columns[:4]:
    colx = list(coordinates[colx].dropna())
    x.append(colx)
    
y = []
for coly in coordinates.columns[4:]:
    coly = list(coordinates[coly].dropna())
    y.append(coly)
    
# now x and y are list of lists containing lons and lats, respectively
````

````python
p = figure(plot_height = 400, plot_width = 400,
           x_axis_label='longitude(degrees)', 
           y_axis_label='latitude(degrees)')

# Add patches to figure p with line_color=white for x and y
p.patches(x,y,line_color='white')

# Specify the name of the output file and show the result
output_notebook()
show(p)
````

### The Bokeh ColumnDataSource

The `ColumnDataSource` is a table-like data object that maps string column names to sequences (columns) of data. It is the central and most common data structure in Bokeh.

````python
# Import the ColumnDataSource class from bokeh.plotting
from bokeh.plotting import ColumnDataSource

# Create a ColumnDataSource: source
source = ColumnDataSource(df)

# Add circle glyphs to the figure p
p.circle(x='hp', y='mpg', size=8, source=source)

#note that there is no need to do df['col'], only 'col'

# Specify the name of the output file and show the result
output_notebook()
show(p)
````

## Customizing glyphs

### Selection and non-selection glyphs

Add **`box_select`** tool to a `figure`, with this tool you can then specify what happens to the circle points you select by the box `selection_` and what happens to the rest `nonselection_`

````python
# Import the ColumnDataSource class from bokeh.plotting
from bokeh.plotting import ColumnDataSource

# Create a ColumnDataSource: source
source = ColumnDataSource(df)

# Create a figure with the "box_select" tool: p
p = figure(x_axis_label='HP',y_axis_label='MPG',tools='box_select')

# Add circle glyphs to the figure p with the selected and non-selected properties
p.circle(x='hp',y='mpg',source=source,selection_color='red',nonselection_alpha=0.1)

# Specify the name of the output file and show the result
output_notebook()
show(p)
````

### Hover glyphs

Hover tools controls what to be displayed when the mouse passes by a data point

* Specify what happens to the color, alpha, .. when the mouse passes by using **`hover_`** in **`p.circle`**
* **`hover = HoverTool(---)`**
    * **`tooltips`** written labels when mouse passes (described in detail in 'annotation' section)
    * **`mode`** 
        * **'mouse'**: only when the mouse is directly over a glyph
        * **'vline'**: whenever the a vertical line from the mouse position intersects a glyph
        * **'hline'**: horizantal line


````python
# import the HoverTool
from bokeh.models import HoverTool

source = ColumnDataSource(df)

# Add a circle glyphs to figure p
p.circle(x='hp',y='mpg',source=source, size=10,
         fill_color='grey', alpha=0.1, line_color=None,
         hover_fill_color='firebrick', hover_alpha=0.5,
         hover_line_color='white')

# Create a HoverTool: hover
hover = HoverTool(tooltips=None, mode='hline')

# Add the hover tool to the figure p 
p.add_tools(hover)

output_notebook()
show(p)
````

### Colormapping

Using the `CategoricalColorMapper`, you can color each glyph by a categorical property.

````python
#Import CategoricalColorMapper from bokeh.models
from bokeh.models import CategoricalColorMapper

df = pd.read_csv('auto-mpg.csv')

p = figure(x_axis_label='Weight',y_axis_label='MPG')

# Convert df to a ColumnDataSource: source
source = ColumnDataSource(df)

# Make a CategoricalColorMapper object: color_mapper
color_mapper = CategoricalColorMapper(factors=['Europe', 'Asia', 'US'],
                                      palette=['red', 'green', 'blue'])

# Add a circle glyph to the figure p
p.circle('weight', 'mpg', source=source,
            color=dict(field='origin', transform=color_mapper),
            legend='origin')

# 'origin' is the name of the column that is colored by

output_notebook()
show(p)
````

## Bokeh Layout

### Rows and Columns of Plots (Subplots)

* **`row`**
* **`column`**
    * **`sizing_mode='scale_width'`**: scale the widths to fill the whole figure

````python
# data

fertility = [1.769,2.682,2.077,2.140, 1.827]
female_literacy = [90.5,50.8,99.0,88.8,90.2]
population = [90,50,33,66,78]
````

````python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show

# Import row from bokeh.layouts
from bokeh.layouts import row
# same with columns

# Create the first figure: p1
p1 = figure(x_axis_label='fertility (children per woman)', 
            y_axis_label='female_literacy (% population)')

# Add a circle glyph to p1
p1.circle(fertility, female_literacy)

# Create the second figure: p2
p2 = figure(x_axis_label='population', 
            y_axis_label='female_literacy (% population)')

# Add a circle glyph to p2
p2.circle(population, female_literacy)

# Put p1 and p2 into a horizontal row: layout
layout = row([p1,p2])

# Specify the name of the output_file and show the result
output_notebook()

# show layout
show(layout)
````

The same can be done using `column`, and nested layouts can be created with different combinations of `row` and `column`

### Gridded Layouts
* **`gridplot`**


````python
# Import gridplot from bokeh.layouts
from bokeh.layouts import gridplot

# define the 3rd plot
p3 = figure(x_axis_label='fertility (children per woman)', 
            y_axis_label='population')
p3.circle(fertility, population)

# Create a gridplot using row1 and row2: layout
layout = gridplot([[p1,p2],[p3,None]])

# Specify the name of the output_file and show the result
output_notebook()
show(layout)
````

### Tabbed Layouts

* **`Panel`**
    * **`child`**: the plot
    * **`title`**: title of the plot

````python
# Import Panel and Tabs from bokeh.models.widgets
from bokeh.models.widgets import Panel,Tabs

# Create tab1 from plot p1: tab1
tab1 = Panel(child=p1, title='Fer vs Lit')

# Create tab2 from plot p2: tab2
tab2 = Panel(child=p2, title='Pop vs Lit')

# Create tab3 from plot p3: tab3
tab3 = Panel(child=p3, title='Fer vs Pop')

# Create a Tabs layout: layout
layout = Tabs(tabs=[tab1, tab2, tab3],)

# Specify the name of the output_file and show the result
output_notebook()
show(layout)
````

## Linking Plots Together

### Linked Axes

Linking axes between plots is achieved by sharing `range` objects.

````python
# Link the x_range of p2 to p1: p2.x_range
p2.x_range = p1.x_range

# Link the y_range of p2 to p1: p2.y_range
p2.y_range = p1.y_range

# Specify the name of the output_file and show the result
output_notebook()
show(layout)

# try zooming one plot, the other one will zoomed accordingly
````

### Linking brushing

By sharing the same `ColumnDataSource` object between multiple plots, selection tools like BoxSelect and LassoSelect will highlight points in both plots that share a row in the ColumnDataSource.

````python
from bokeh.plotting import ColumnDataSource

df = pd.read_csv('auto-mpg.csv')
source = ColumnDataSource(df)

p1 = figure(x_axis_label='Weight',y_axis_label='MPG', 
            tools='box_select,lasso_select')
p1.circle('weight', 'mpg',source=source)

p2 = figure(x_axis_label='HP',y_axis_label='MPG',
            tools='box_select,lasso_select')
p2.circle('hp', 'mpg', source=source)

layout = row(p1, p2)

output_notebook()
show(layout)
````

Try using the `box_select` or the `lasso_select` in one plot and notice what happens to the other

## Bokeh Server

The main purpose of the Bokeh server is to synchronize python objects with web applications in a browser, so that rich, interactive data applications can be connected to powerful PyData libraries such as NumPy, SciPy, Pandas, and scikit-learn.

### Bokeh Apps
1. Create plots and wedgets
2. Add callbacks: functions that perform certain operations when certain action happens
3. Arrange plots and wedgets in layouts
4. **`curdoc().add_root()`**
5. **`bokeh serve file_name.py`** in a command line

### Add a single plot to the app

````python
# Perform necessary imports
from bokeh.io import curdoc
from bokeh.plotting import figure

# Create a new plot: plot
plot = figure()

# Add a line to the plot
plot.line(x=[1,2,3,4,5],y=[2,5,4,6,7])

# Add the plot to the current document
curdoc().add_root(plot)
output_notebook()
show(plot)
````

### Add a single slider

````python
# Perform the necessary imports
from bokeh.io import curdoc
from bokeh.layouts import widgetbox
from bokeh.models import Slider

# Create a slider: slider
slider = Slider(title='my slider', start=0, end=10, step=0.1, value=2) #initial value=2

# Create a widgetbox layout: layout
layout = widgetbox(slider)

# Add the layout to the current document
curdoc().add_root(layout)
output_notebook()
show(layout)
````

### Muliple sliders in one document

````python
# Create first slider: slider1
slider1 = Slider(title='slider1',start=0,end=10,step=0.1,value=2)

# Create second slider: slider2
slider2 = Slider(title='slider2',start=10,end=100,step=1,value=20)

# Add slider1 and slider2 to a widgetbox
layout = widgetbox(slider1,slider2)

# Add the layout to the current document
curdoc().add_root(layout)
output_notebook()
show(layout)
````

### Combine Bokeh models into layouts, and connect them using callbacks

* **`def callback(attr,old,new)`** the function that updates the wedget
* **`slider.on_change`** to connect silder with the function callback that determines how slide would change with the plot

The code below can be copied into a Python file of your name; let's say `bokehapp.py`. Copy it and paste it into the file, then use Windows Command Prompt and run 

```bash
bokeh serve --show bokehapp.py
```

NOTE that you might want to install `bokeh` package or any other package like `pandas`. Just use:
```bash
pip install ...
```

Replace `...` with the package you want to install

```python
# bokehapp.py

from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook

from bokeh.plotting import ColumnDataSource
from bokeh.layouts import row, column

from bokeh.io import curdoc
from bokeh.layouts import widgetbox
from bokeh.models import Slider

import numpy as np
import pandas as pd


#-----------------------
# Create ColumnDataSource: source
N=300
source = ColumnDataSource(data={'x':np.random.random(N),
                                'y':np.random.random(N)})
# 300 random numbers from 0 to 1

# Add a line to the plot
plot=figure()
plot.circle(x='x', y='y', source=source)

# add a slider
slider = Slider(start=100,end=1000,value=N,step=10,title='Number of Points')

# Define a callback function: callback
def callback(attr, old, new): #always this way

    # Read the current value of the slider: scale
    scale = slider.value

    # Update source with the new data values
    source.data = {'x': np.random.random(scale), 
                   'y': np.random.random(scale)}

# Attach the callback to the 'value' property of slider
slider.on_change('value',callback)

# Create a column layout: layout
layout = column(widgetbox(slider), plot) # the silder that is created 2 exercises ago
# calling the eidgetbox first in column to make sure slider is above plot

curdoc().add_root(layout)
```

### Updating plots from dropdowns

* **`from bokeh.plotting import Select`**

```python
# bokehapp.py

from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook

from bokeh.plotting import ColumnDataSource
from bokeh.layouts import row, column

from bokeh.io import curdoc
from bokeh.layouts import widgetbox
from bokeh.models import Slider,Select

from numpy.random import random, normal, lognormal
import pandas as pd


#-----------------------
# Create ColumnDataSource: source
N=300
source = ColumnDataSource(data={'x':random(N),
                                'y':random(N)})
# 300 random numbers from 0 to 1

# Add a line to the plot
plot=figure()
plot.circle(x='x', y='y', source=source)

# add a slider
menu = Select(options=['unifrom','normal','lognormal'],value='uniform',
             title='Distribution')
# value = current selection from the menu (in options)

# Define a callback function: callback
def callback(attr, old, new): #always this way
    
    if   menu.value == 'uniform': f = random
    elif menu.value ==  'normal': f = normal
    else:                         f = lognormal
    source.data={'x':f(size=N),'y':f(size=N)}
menu.on_change('value',callback)

# Create a column layout: layout
layout = column(menu, plot)
```

### Updating data sources from dropdown callbacks

```python
# data

fertility = [1.769,2.682,2.077,2.140, 1.827]
female_literacy = [90.5,50.8,99.0,88.8,90.2]

    'x' : fertility,
    'y' : female_literacy
})

plot.circle('x', 'y', source=source)

# Define a callback function: update_plot
def update_plot(attr, old, new):
    # If the new Selection is 'female_literacy', update 'y' to female_literacy
    if new == 'female_literacy': 
        source.data = {
            'x' : fertility,
            'y' : female_literacy
        }
    # Else, update 'y' to population
    else:
        source.data = {
            'x' : fertility,
            'y' : population
        }

```

### Synchronize two dropdowns

```python
        select2.value = '100'
```

### Buttons

The basic code for importing Buttons in your Bokeh plots is much simpler than the other widgets, their callbacks don't need any input arguments. The strucute is as follows:

```python
from bokeh.models import Button

button = Button(label='press me')

def update(): # no argument
    
    # Do something!
button.on_click(update)
```

**Example of Button widgets**
```python
# Create a Button with label 'Update Data'
button = Button(label='Update Data')

# Define an update callback with no arguments: update
def update():

    # Compute new y values: y
    y = np.sin(x) + np.random.random(N)

    # Update the ColumnDataSource data dictionary
    source.data = {'x':x,
        'y':y
    }

# Add the update callback to the button
button.on_click(update)

# Create layout and add to current document
layout = column(widgetbox(button), plot)
curdoc().add_root(layout)
```

```python
# Read file
import pandas as pd

data.head()

from bokeh.plotting import figure

# Define the callback: update_plot
def update_plot(attr, old, new):
    # Read the current value off the slider and 2 dropdowns: yr, x, y
    yr = slider.value #slider value
    x = x_select.value #first dropdwon
    y = y_select.value #second dropdown
    # Label axes of plot
    plot.xaxis.axis_label = x
    plot.yaxis.axis_label = y

    plot.x_range.end = max(data[x])
    plot.y_range.start = min(data[y])
    plot.y_range.end = max(data[y])

    plot.title.text = 'Gapminder data for %d' % yr

# Create a dropdown slider widget: slider

    title='y-axis data'
)
```
