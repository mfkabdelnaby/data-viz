# Seaborn

A library for Statistical Data Visualization and works better with pandas df

````python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
````

## Plotting Histograms in Seaborn

```python
sns.set()
plt.hist(df['col'])
```

## Simple Linear Regressions 

* **`sns.lmplot(x='xcol',y='ycol',data=df)`**

````python
auto = pd.read_csv('auto-mpg.csv')
auto.head()
````

````python
# Plot a linear regression between 'weight' and 'hp'
sns.lmplot(x='weight', y='hp', data=auto)

plt.style.use('seaborn')
plt.show()
````

You can see there is a STRONG correlation betweeb `weight` and `hp`

## Plotting residuals of a regression

Often you want to see the *residuals of a regression* to get a better idea how well the regression captured the data

* **`sns.residplot()`**

````python
# Generate a green residual plot of the regression between 'hp' and 'mpg'
sns.residplot(x='hp', y='mpg', data=auto, color='green')

# Display the plot
plt.show()
````

## Higher-order regressions

When there are more complex relationships between two variables, a simple first order regression is often not sufficient to accurately capture the relationship between the variables. 

* **`sns.regplot()`**
    * **`order`** to control the order of polynomial regression.
    * **`scatter=None`** to prevent plotting the scatter plot points again.
    * **`color`** to define the color like matplotlib
    * **`label`**

````python
# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None, color='blue', label='order 1')

# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None, order=2, color='green', label='order 2')

# Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()
````

It seems that 2nd order regression is necessary to capture the relationship between `weight` and `mpg`

### Grouping linear regressions by hue

In **`sns.lmplot`:**
* **`hue`**: column you want to groupby your regression
* **`palette`** for coloring distinct groups

````python
# Plot a linear regression between 'weight' and 'hp', with a hue of 'origin' and palette of 'Set1'
sns.lmplot(x='weight',y='hp',data=auto,hue='origin',palette='Set1')

# Display the plot
plt.show()
````

### Grouping linear regressions by row or column
Rather than overlaying linear regressions of grouped data in the same plot, we may want to use a grid of subplots

In **`sns.lmplot`:**
* **`row`**
* **`col`**

````python
sns.lmplot(x='weight',y='hp',data=auto,hue='origin',
           palette='Set1',row='origin')
plt.show()
````

## Visualizing univariate distributions

Regressions are useful to understand relationships between two continuous variables. Often we want to explore how the distribution of a single continuous variable is affected by a second categorical variable

### Strip plots

* **`sns.stripplot(x='groupedby',y='col',data=df)`**

    * **`jitter=True`** to spreadout overlapping points
    * **`size`** decrease the point size

Other matplotlib arguments can be passed to `sns.stripplot()`, e.g., `marker`, `color`, `size`, etc.

````python
# Make a strip plot of 'hp' grouped by 'cyl'
plt.subplot(2,1,1)
sns.stripplot(x='cyl', y='hp', data=auto)

# Make the strip plot again using jitter and a smaller point size
plt.subplot(2,1,2)
sns.stripplot(x='cyl', y='hp', data=auto,jitter=True,size=3)

# Display the plot
plt.show()
````

### Swarm Plots
* **`sns.swarmplot()`**
    * Similar sintax as `sns.stripplot`
    * Orientation of `x` and `y` can be interred from the choice of the columns
    * **`orient='h'/'v'`** to force orientation
    * **`hue`** another grouping by coloring the points for each group

````python
# Generate a swarm plot of 'hp' grouped horizontally by 'cyl'  
plt.subplot(2,1,1)
sns.swarmplot(x='cyl', y='hp', data=auto)

# Generate a swarm plot of 'hp' grouped vertically by 'cyl' with a hue of 'origin'
plt.subplot(2,1,2)
sns.swarmplot(x='hp', y='cyl', data=auto,hue='origin',
              orient='h')

# Display the plot
plt.show()
````

### Violin Plot

* **`sns.violinplot()`**
    * Same parameters as above
    * **`inner=None`** to remove the inner annotation, usually you need to remove them when adding a `stripplot` inside it

````python
# Generate a violin plot of 'hp' grouped horizontally by 'cyl'
plt.subplot(2,1,1)
sns.violinplot(x='cyl', y='hp', data=auto)

# Generate the same violin plot again with a color of 'lightgray' and without inner annotations
plt.subplot(2,1,2)
sns.violinplot(x='cyl', y='hp', data=auto,inner=None,color='lightgray')

# Overlay a strip plot on the violin plot
sns.stripplot(x='cyl', y='hp', data=auto,jitter=True,size=1.5)

# Display the plot
plt.show()
````

## Visualizing multivariate distributions

### Joint distributions

* **`sns.jointplot()`**


````python
# Generate a joint plot of 'hp' and 'mpg'
sns.jointplot(x='hp',y='mpg',data=auto)

# Display the plot
plt.show()
````

**`kind`** parameter in `sns.jointplot()`:
* **`kind='scatter'`** scatter plot
* **`kind='reg'`** regression plot of order1 (default)
* **`kind='resid'`** residual plot
* **`kind='kde'`** kernel density estimate
* **`kind='hex'`** hexbin plot

````python
# Generate a joint plot of 'hp' and 'mpg' using a kde plot
sns.jointplot(x='hp',y='mpg',data=auto,kind='kde')

# Display the plot
plt.show()
````

### Plotting distributions pairwise

* **`sns.pariplot(df)`**

Constructs a grid of all joint plots pairwise from all pairs of (non-categorical) columns in a DataFrame

````python
auto1 = auto[['mpg','hp','cyl','origin']]
auto1.head()
````

````python
# Plot the pairwise joint distributions from the DataFrame 
sns.pairplot(auto1)
plt.show()
````

To add regression plots on top of the scatter:
* **`kind='reg`**

To add a colored groupby use:
* **`hue`**

````python
# Plot the pairwise joint distributions grouped by 'origin' along with regression lines
sns.pairplot(auto1,kind='reg',hue='origin')
plt.show()
````

### Heatmap
A heatmap is useful to compute covariances between the variables instead. Its diagonal entries are all 1s, and the off-diagonal entries are between -1 and +1 (quantifying the degree to which variable pairs vary jointly)

* **`sns.heatmap()`**

````python
import numpy as np
import pandas as pd

data = pd.DataFrame(np.random.rand(11, 5), 
                    columns=['A', 'B', 'C', 'D', 'E'], 
                    index = range(2000, 2011, 1))
data
````

````python
sns.heatmap(data) #the data is the correlation here
````

For your dataframe, the heatmap between its columns can be found by:

```python
sns.heatmap(data.corr(), square=True, cmap='RdYlGn')
```
